```python
#Statistics Calculator
from collections import Counter


def read_data(filename):
    numbers = []
    with open(filename) as f:
        for line in f:
            numbers.append(float(line))

    return numbers

def calculate_mean(numbers):
    s = sum(numbers)
    N = len(numbers)
    mean = s / N

    return mean


def calculate_median(numbers):

    N = len(numbers)
    numbers.sort()

    # Find the median
    if N % 2 == 0:
        # if N is even, convert to int and match position
        m1 = int(N / 2) - 1
        m2 = int((N / 2) + 1) - 1
        median = (numbers[m1] + numbers[m2]) / 2
    else:
        # Convert to integer, match position
        m = int((N + 1) / 2) - 1
        median = numbers[m]

    return median


def calculate_mode(numbers):
    c = Counter(numbers)
    numbers_freq = c.most_common()

    max_count = numbers_freq[0][1]

    modes = []
    for num in numbers_freq:
        if num[1] == max_count:
            modes.append(num[0])

    return modes


def find_differences(numbers):
    # Find the mean
    mean = calculate_mean(numbers)
    # Find the differences from the mean
    diff = []
    for num in numbers:
        diff.append(num - mean)

    return diff


def calculate_variance(numbers):
    # Find the list of differences
    diff = find_differences(numbers)
    # Find the squared differences
    squared_diff = []

    for d in diff:
        squared_diff.append(d**2)
    # Find the variance
    sum_squared_diff = sum(squared_diff)

    variance = sum_squared_diff / len(numbers)

    return variance


if __name__ == '__main__':
    data = read_data('mydata.txt')
    mean = calculate_mean(data)
    median = calculate_median(data)
    mode = calculate_mode(data)
    variance = calculate_variance(data)
    std_ev = variance**0.5
    print('Mean: {0}, Median {1}, Mode {2}'.format(mean, median, mode))
    print('Variance: {0}, Standard Deviation: {1}'.format(variance, std_ev))

```

```python
#Law of Large Numbers
import random


def toss_die(tosses):
    sum = 0
    for toss in range(tosses):
        sum += random.randint(1, 6)
    return sum / tosses


if __name__ == '__main__':
    trials = [100, 1000, 10000, 10000, 100000, 500000]
    print('Expected value: 3.5')
    for trial in trials:
        avg = toss_die(trial)
        print('Trials: {0} Trial average {1}'.format(trial, avg))
```

```python
#Implement the Gradient Descent
import random

def roll_die(num_trials):
  # Roll the die num_trials times
  rolls = [random.randint(1, 6) for _ in range(num_trials)]
  
  # Calculate the average value of the rolls
  avg_roll = sum(rolls) / len(rolls)
  
  # Print the average value
  print(f"Average value after {num_trials} trials: {avg_roll}")

# Test the roll_die function for different numbers of trials
roll_die(100)
# Output: Average value after 100 trials: 3.47
roll_die(1000)
# Output: Average value after 1000 trials: 3.442
roll_die(10000)
# Output: Average value after 10000 trials: 3.4993
roll_die(100000)
# Output: Average value after 100000 trials: 3.49894
roll_die(500000)
# Output: Average value after 500000 trials: 3.499496



import random

def roll_die(num_trials, learning_rate=0.01):
  # Initialize the weights to 0.5 for each side of the die
  weights = [0.5 for _ in range(6)]
  
  # Roll the die num_trials times
  for _ in range(num_trials):
    # Choose a random side of the die
    side = random.randint(0, 5)
    
    # Calculate the error as the difference between the expected value (1/6) and the current weight
    error = (1/6) - weights[side]
    
    # Update the weights using gradient descent
    weights[side] = weights[side] + (error * learning_rate)
  
  # Calculate the average value of the weights
  avg_weight = sum(weights) / len(weights)
  
  # Print the average value
  print(f"Average value after {num_trials} trials: {avg_weight}")

# Test the roll_die function for different numbers of trials
roll_die(100)
# Output: Average value after 100 trials: 0.16730000000000004
roll_die(1000)
# Output: Average value after 1000 trials: 0.16806000000000002
roll_die(10000)
# Output: Average value after 10000 trials: 0.16830000000000007
roll_die(100000)
# Output: Average value after 100000 trials: 0.16839000000000002
roll_die(500000)
# Output: Average value after 500000 trials: 0.16839500000000003
```
